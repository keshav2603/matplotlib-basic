```python
import matplotlib.pyplot as plt
import numpy as np
```

### line plot

```python
x= np.array([1,2,3,4,5])
y=np.array([10,20,40,60,90])
x1=np.array([2,3,5,6,8])
y1=np.array([10,30,20,50,70])
plt.plot(x,y,label="model A")
plt.plot(x1,y1, label="Model B")
plt.title("Simple Line Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.grid(True)
plt.legend()
plt.show()
```

```python
epochs = [1, 2, 3, 4, 5]
train_loss = [0.9, 0.6, 0.4, 0.3, 0.25]
val_loss = [1.0, 0.7, 0.5, 0.4, 0.38]

plt.plot(epochs, train_loss, label='Train Loss', color='blue')
plt.plot(epochs, val_loss, label='Validation Loss', color='orange')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(
    x,                # x-axis values
    y,                # y-axis values
    color='blue',     # line color
    marker='o',       # point marker (like 'o', 's', '^')
    linestyle='--',   # line style ('-', '--', '-.', ':', '')
    linewidth=2,      # line thickness
    markersize=6,     # size of the marker
    alpha=0.8,        # transparency (0 = invisible, 1 = solid)
    label='My Line',  # legend label
    markerfacecolor='red',  # marker fill color
    markeredgecolor='black',  # marker border color
    markeredgewidth=1.5,      # marker border width
)
plt.title("Simple Line Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.grid(True)
plt.legend()
plt.show()
```

# scatter plot

```python
x=np.array([1,2,4,5,6,7,8,9])
y=np.array([10,20,30,40,50,60,40,20])
plt.scatter(
    x, y,
    color='blue',         # point color
    s=100,                # size of points
    alpha=0.7,            # transparency
    marker='*',           # shape of points
    edgecolor='black',    # border color
    label='Data points'   # for legend
)
plt.title("Basic Scatter Plot")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python

y_true = np.array([100, 120, 140, 160])
y_pred = np.array([110, 115, 135, 170])

plt.scatter(y_true, y_pred, color='green', s=80)
# plt.plot([100, 160], [100, 160], color='red', linestyle='--', label='Perfect Prediction')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Predicted vs Actual")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
labels = np.random.randint(0, 2, 50)

plt.scatter(x, y, c=labels, cmap='coolwarm')
plt.title("2D Data Clustering")
plt.xlabel("Feature X")
plt.ylabel("Feature Y")
plt.grid(True)
plt.show()
```

# Bar plot


```python
x=["c++", "python", "javascript", "typescript"]
y=[10,80,30,40]
plt.bar(x,y)
plt.title("basic bar plot")
plt.xlabel("category")
plt.ylabel("value")
plt.show()
```

```python
plt.bar(
    x=['A', 'B', 'C'],             # categories
    height=[10, 20, 15],           # values
    width=0.6,                     # bar width
    color=["red"],              # bar fill color
    edgecolor='yellow',            # bar border color
    linewidth=1.5,                # border width
    align='center',               # alignment: 'center' or 'edge'
    alpha=0.8                     # transparency (0 to 1)
)
```

# HISTOGRAM

```python
data = np.random.randn(1000)  # 1000 normal distributed values

plt.hist(data, bins=20, color='orange', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(
    x=data,                       # input data (1D)
    bins=20,                      # number of bins
    color=["orange","yellow"],               # fill color
    edgecolor='black',            # border color
    linewidth=1.2,                # border width
    alpha=0.7,                    # transparency
    histtype='bar',               # bar, step, stepfilled
    orientation='vertical',       # vertical or horizontal
    cumulative=False              # True to show cumulative
)
```

# Box Plot

```python
data = [np.random.normal(60, 5, 100), np.random.normal(50, 10, 100)]
plt.boxplot(
    x=data,                       # list or 2D array
    notch=False,                  # notch shape at median
    vert=True,                    # vertical or horizontal
    patch_artist=True,           # fill boxes with color
    widths=0.5,                   # width of boxes
    boxprops=dict(color='blue', facecolor='lightblue'),
    medianprops=dict(color='red'),
    whiskerprops=dict(color='black'),
    capprops=dict(color='green'),
    flierprops=dict(marker='o', color='purple', alpha=0.5))
plt.show()
```

# subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
#   This is subplot number 1 in a 2x2 grid
plt.subplot(2, 2, 1)  # This is subplot number 1 in a 2x2 grid


plt.show()


```

```python
# Create dummy data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x ** 2
y4 = np.random.rand(100)

# Create subplots: 2 rows, 2 columns
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

# Top-left: Line plot
axes[0, 0].plot(x, y1, color='blue')
axes[0, 0].set_title("Sine Wave")

# Top-right: Line plot
axes[0, 1].plot(x, y2, color='green')
axes[0, 1].set_title("Cosine Wave")

# Bottom-left: Scatter plot
axes[1, 0].scatter(x, y4, color='red')
axes[1, 0].set_title("Random Scatter")

# Bottom-right: Bar plot
axes[1, 1].bar([1, 2, 3, 4], [10, 20, 15, 5])
axes[1, 1].set_title("Bar Chart")

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

# fill between

```python
plt.fill_between(
    x,                             # X values
    y1,                            # Lower bound
    y2,                            # Upper bound (optional)
    color='lightblue',
    alpha=0.5,
    label='Confidence Range',
    interpolate=True               # smooth if NaN in y
)
```
