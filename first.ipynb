```python
import matplotlib.pyplot as plt
import numpy as np
```

### line plot

```python
x= np.array([1,2,3,4,5])
y=np.array([10,20,40,60,90])
x1=np.array([2,3,5,6,8])
y1=np.array([10,30,20,50,70])
plt.plot(x,y,label="model A")
plt.plot(x1,y1, label="Model B")
plt.title("Simple Line Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.grid(True)
plt.legend()
plt.show()
```

```python
epochs = [1, 2, 3, 4, 5]
train_loss = [0.9, 0.6, 0.4, 0.3, 0.25]
val_loss = [1.0, 0.7, 0.5, 0.4, 0.38]

plt.plot(epochs, train_loss, label='Train Loss', color='blue')
plt.plot(epochs, val_loss, label='Validation Loss', color='orange')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(
    x,                # x-axis values
    y,                # y-axis values
    color='blue',     # line color
    marker='o',       # point marker (like 'o', 's', '^')
    linestyle='--',   # line style ('-', '--', '-.', ':', '')
    linewidth=2,      # line thickness
    markersize=6,     # size of the marker
    alpha=0.8,        # transparency (0 = invisible, 1 = solid)
    label='My Line',  # legend label
    markerfacecolor='red',  # marker fill color
    markeredgecolor='black',  # marker border color
    markeredgewidth=1.5,      # marker border width
)
plt.title("Simple Line Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.grid(True)
plt.legend()
plt.show()
```
