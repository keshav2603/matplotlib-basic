# relational plot

```python
tips=sns.load_dataset("tips")
tips.sample(10)
```

```python
#scatter plot -> axes level function
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time",size="size")
```

```python
# scatter using relplot() -> figure level function
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter",hue="sex", style="time",size="size")
```

```python
gap= px.data.gapminder()
gap_india=gap[gap["country"]=="India"]
gap_india.sample(10)
```

```python
#line plot

sns.relplot(data=gap_india, x="year", y="lifeExp",kind="line",) 
```

```python
gap_asia= gap[gap["country"].isin(["India", "Germany", "Brazil"])]
gap_asia.sample(10)
```

```python
sns.relplot(data=gap_asia, x="year", y="lifeExp",kind="line",hue="country",style="continent") 
```

```python
#facet plot -> work only with figure level function
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter",col="sex",row="smoker", hue="time")
# sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter",col="sex",row="day", hue="time")
```

```python
gap
sns.relplot(gap,x="lifeExp", y="gdpPercap",kind="scatter",col="year",col_wrap=4)
```
