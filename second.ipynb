# relational plot

```python
tips=sns.load_dataset("tips")
tips.sample(10)
```

```python
#scatter plot -> axes level function
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time",size="size")
```

```python
# scatter using relplot() -> figure level function
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter",hue="sex", style="time",size="size")
```

```python
gap= px.data.gapminder()
gap_india=gap[gap["country"]=="India"]
gap_india.sample(10)
```

```python
#line plot

sns.relplot(data=gap_india, x="year", y="lifeExp",kind="line",) 
```

```python
gap_asia= gap[gap["country"].isin(["India", "Germany", "Brazil"])]
gap_asia.sample(10)
```

```python
sns.relplot(data=gap_asia, x="year", y="lifeExp",kind="line",hue="country",style="continent") 
```

```python
#facet plot -> work only with figure level function
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter",col="sex",row="smoker", hue="time")
# sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter",col="sex",row="day", hue="time")
```

```python
gap
sns.relplot(gap,x="lifeExp", y="gdpPercap",kind="scatter",col="year",col_wrap=4)
```

# matric plot 
heatmap

```python
temp=gap[gap["continent"]=="Europe"].pivot(index="country", columns=["year"], values=["lifeExp"])
```

```python
# heatmap is axis level only
plt.figure(figsize=(15,15))
sns.heatmap(data=temp,annot=True,linewidths=0.5,cmap="winter")
```

# categorical plot

```python
# strip plot
tips
# sns.stripplot(data=tips, x="day", y="total_bill",jitter=False)
sns.stripplot(data=tips, x="day", y="total_bill")
```

```python
sns.catplot(data=tips,kind="strip",x="day",y="total_bill")
```

```python
# jitter
sns.catplot(data=tips,kind="strip",x="day",y="total_bill",hue="sex",jitter=0.2)
```

```python
# swarmplot
sns.catplot(kind="swarm",data=tips, y="total_bill",x="day",hue='sex')
```

```python
#boxplot
# sns.catplot(kind="box",data=tips,x="sex",y="total_bill")
sns.boxplot(data=tips,x="sex",y="total_bill")
```

```python
#volinplot
# sns.catplot(data=tips,kind="violin",y="total_bill",x="day")
sns.violinplot(data=tips,y="total_bill",x="day",hue="sex",split=True)
```

```python
#barplot
# sns.barplot(data=tips,x="sex",y="total_bill")
sns.catplot(data=tips,x="sex",y="total_bill",kind="bar",hue="smoker",estimator=np.median)

```

```python
# point plot
sns.catplot(data=tips,x="sex",y="total_bill",kind="point",hue="smoker",estimator=np.median,ci=None)
```

```python
#faceting using catplot
sns.catplot(data=tips, x="sex",y="total_bill", col="smoker",kind="strip",row="day")
```

```python
sns.residplot(data=tips,x="total_bill", y="tip")
```

# multigridplot

```python
#facetgrid

g = sns.FacetGrid(data=tips, col="day", row="time",hue="smoker")
g.map(sns.scatterplot,"tip","total_bill")
g.add_legend()
```

# Pairplot

```python
iris
```

```python
sns.pairplot(data=iris,hue="species")
```

```python
#pairgrid
g=sns.PairGrid(data=iris,hue="species")
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris,hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

# joinplot

```python
sns.jointplot(data=tips,x="total_bill",y="tip")
```

```python
# jointgrid
g=sns.JointGrid(data=tips, x="total_bill",y="tip")
g.plot(sns.scatterplot,sns.histplot)
```
